### Download Additional Images

```python
def bar_progress(current, total, width=80):
    progress_message = "Downloading: %d%% [%d / %d] bytes" % (current / total * 100, current, total)
    # Don't use print() as it will print in new line every time.
    sys.stdout.write("\r" + progress_message)
    sys.stdout.flush()

# DOWNLOAD DATASETS
if not os.path.isdir('datasets'):
    os.mkdir('datasets')

# Human-Interaction-Images
site_url = 'https://vision.cs.hacettepe.edu.tr/interaction_images/Facial_Descriptors_files/facial_descriptors_dataset.zip'
file_name = wget.download(site_url, out='datasets/human-interaction.zip', bar=bar_progress)
with zipfile.ZipFile(file_name, 'r') as zip_ref:
    zip_ref.extractall('datasets/human-interaction/')

# SBU-Kinect-Interaction
site_url = ['http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s01s02.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s01s03.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s01s07.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s02s01.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s02s03.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s02s06.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s02s07.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s03s02.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s03s04.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s03s05.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s03s06.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s04s02.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s04s03.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s04s06.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s05s02.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s05s03.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s06s02.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s06s03.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s06s04.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s07s01.zip',
            'http://vision.cs.stonybrook.edu/~kiwon/Datasets/SBU_Kinect_Interactions/s07s03.zip']
for url in site_url:
    file_name = wget.download(url, out='datasets/SBU-Kinect-Interaction.zip', bar=bar_progress)
    with zipfile.ZipFile(file_name, 'r') as zip_ref:
        zip_ref.extractall('datasets/SBU-Kinect-Interaction/')

# TV-Human-interaction
site_url = "http://www.robots.ox.ac.uk/~alonso/data/tv_human_interactions_videos.tar.gz"
file_name = wget.download(site_url, out='datasets/TV-Human-interaction.tar.gz', bar=bar_progress)
tar = tarfile.open(file_name, "r:gz")
tar.extractall(path='datasets/TV-Human-interaction')
tar.close()

# ShakeFive2
site_url = "http://www2.projects.science.uu.nl/shakefive/data/ShakeFive2.videos.tar"
file_name = wget.download(site_url, out='datasets/ShakeFive2.tar', bar=bar_progress)
tar = tarfile.open(file_name, "r:")
tar.extractall(path='datasets/ShakeFive2/')
tar.close()

# UT-Interaction
site_url = "http://cvrc.ece.utexas.edu/SDHA2010/videos/competition_1/ut-interaction_set1.zip"
file_name = wget.download(site_url, out='datasets/UT-Interaction_set1.zip', bar=bar_progress)
with zipfile.ZipFile(file_name, 'r') as zip_ref:
    zip_ref.extractall('datasets/UT-Interaction/ut-interaction_set1')

site_url = "http://cvrc.ece.utexas.edu/SDHA2010/videos/competition_1/ut-interaction_set2.zip"
file_name = wget.download(site_url, out='datasets/UT-Interaction_set2.zip', bar=bar_progress)
with zipfile.ZipFile(file_name, 'r') as zip_ref:
    zip_ref.extractall('datasets/UT-Interaction/ut-interaction_set2')

# VOC dataset
site_url = 'http://host.robots.ox.ac.uk/pascal/VOC/voc2012/VOCtrainval_11-May-2012.tar'
file_name = wget.download(site_url, out='datasets/VOC.tar', bar=bar_progress)
tar = tarfile.open(file_name, "r:")
tar.extractall(path='datasets/VOC/')
tar.close()

# COCO dataset
site_url = "http://images.cocodataset.org/zips/train2017.zip"
file_name = wget.download(site_url, out='datasets/Coco_train.zip', bar=bar_progress)
with zipfile.ZipFile(file_name, 'r') as zip_ref:
    zip_ref.extractall('datasets/COCO/')
site_url = "http://images.cocodataset.org/zips/val2017.zip"
file_name = wget.download(site_url, out='datasets/Coco_val.zip', bar=bar_progress)
with zipfile.ZipFile(file_name, 'r') as zip_ref:
    zip_ref.extractall('datasets/COCO/')
    
cocotrain_images_id = [int(im.split('/')[-1].split('.')[0]) for im in glob.glob('datasets/COCO/train2017/*.jpg')]
cocoval_images_id = [int(im.split('/')[-1].split('.')[0]) for im in glob.glob('datasets/COCO/val2017/*.jpg')]

# Load H2O images metadata
metadata_file = "metadata.json"
train_images = "images_trainval.ids"
test_images = "images_test.ids"

train_imageslist = open(train_images, 'r').read().split('\n')[:-1]
test_imageslist = open(test_images, 'r').read().split('\n')[:-1]

with open(metadata_file, 'r') as f:
     metadata = json.load(f)

if os.path.isdir('H2O'):
    rmtree('H2O')

os.mkdir('H2O')
os.mkdir('H2O/trainval_images')
os.mkdir('H2O/test_images')
```

### Parse out images that match train_imageslist and test_imageslist and copy over

```python
for d in metadata['images']:
    print(d['File name'])
    filename = 'HH' + d['File name'].split('.')[0].zfill(10) + '.jpg'
    if filename.split('.')[0] in train_imageslist:
        final_destination = 'H2O/trainval_images/' + filename
    elif filename.split('.')[0] in test_imageslist:
        final_destination = 'H2O/test_images/' + filename

    if d['Url'].endswith('jpg') or d['Url'].endswith('JPG') or d['Url'].endswith('jpeg') or d['Url'].endswith('png')\
            or d['Url'].endswith('ashx'):
        headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"}
        try:
            r = requests.get(d['Url'], stream=True, headers=headers)
            if r.status_code == 200:
                with open(final_destination, 'wb') as f:
                    r.raw.decode_content = True
                    shutil.copyfileobj(r.raw, f)
                try:
                    image = Image.open(final_destination)
                except:
                    os.remove(final_destination)
                    file = wget.download(d['Url'], out=final_destination)
                    try:
                        image = Image.open(final_destination)
                    except:
                        print('Url not available anymore:', d['Url'])
            else:
                if "MS-COCO" in d['Source']:
                    if d['MS-COCO Id'] in cocotrain_images_id:
                        original_source = 'datasets/COCO/train2017/' + str(d['MS-COCO Id']).zfill(12) + '.jpg'
                        copyfile(original_source, final_destination)
                    elif d['MS-COCO Id'] in cocoval_images_id:
                        original_source = 'datasets/COCO/val2017/' + str(d['MS-COCO Id']).zfill(12) + '.jpg'
                        copyfile(original_source, final_destination)
                else:
                    print('Url not available anymore:', d['Url'])
        except:
            print('image ' + d['Url'] + ' is broken')

    elif "Human-Interaction-Images" in d['Source']:
        original_source = 'datasets/human-interaction/facial_descriptors_dataset/' + d['Original file name']
        copyfile(original_source, final_destination)
    elif "SBU-Kinect-Interaction" in d['Source']:
        original_source = 'datasets/SBU-Kinect-Interaction/' + d['Original file name']
        copyfile(original_source, final_destination)
    elif "TV-Human-interaction" in d['Source']:
        video_path = 'datasets/TV-Human-interaction/tv_human_interactions_videos/' + d['Original file name']
        cap = cv2.VideoCapture(video_path)
        frame = 0
        while(cap.isOpened()):
            ret, im_dataset = cap.read()
            if ret:
                if frame == d['Frame Id']:
                    cv2.imwrite(final_destination, im_dataset)
                    break
                frame += 1
    elif "ShakeFive2" in d['Source']:
        video_path = 'datasets/ShakeFive2/' + d['Original file name']
        cap = cv2.VideoCapture(video_path)
        frame = 0
        while(cap.isOpened()):
            ret, im_dataset = cap.read()
            if ret:
                if frame == d['Frame Id']:
                    im_dataset = cv2.resize(im_dataset, (1280, 738), interpolation=cv2.INTER_AREA)
                    cv2.imwrite(final_destination, im_dataset)
                    break
                frame += 1
    elif "UT-Interaction" in d['Source']:
        video_path = 'datasets/UT-Interaction/' + d['Original file name']
        cap = cv2.VideoCapture(video_path)
        frame = 0
        while(cap.isOpened()):
            ret, im_dataset = cap.read()
            if ret:
                if frame == d['Frame Id']:
                    cv2.imwrite(final_destination, im_dataset)
                    break
                frame += 1
    elif "Pascal-Voc" in d['Source']:
        original_source = 'datasets/VOC/VOCdevkit/VOC2012/JPEGImages/' + d['Original file name']
        copyfile(original_source, final_destination)

for d in glob.glob('BIT_images/*.jpg'):
    name = d.split('/')[-1]
    if name.split('.')[0] in train_imageslist:
        copyfile(d, 'H2O/trainval_images/' + name)
    elif name.split('.')[0] in test_imageslist:
        copyfile(d, 'H2O/test_images/' + name)

# Delete downloaded datasets
#rmtree('datasets')
```

### Copy images to train/test directories and rename
#### Files were renamed HO[vcoco_id.zfill(10)].jpg.

```python
# I ran this seperately for the val2014 directory and the train2014 directorry
for image in glob.glob('../vcoco/mscoco2014/val2014/*.jpg'):
    image2 = image.split('/')[-1]
    image3 = image2.split('_')[-1]
    image4 = image3.split('.')[0]
    image5 = int(image4)
    if str(image5) in trainval_imageslist:
        final_source = '../vcoco/mscoco2014/trainval_images/' + 'HO' + str(image5).zfill(10) + '.jpg'
        copyfile(image,final_source) 
    elif str(image5) in test_imageslist:
        final_source = '../vcoco/mscoco2014/test_images/' + 'HO' + str(image5).zfill(10) + '.jpg'
        copyfile(image,final_source) 
```
